### Import bibliotek

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

### Wykres liniowy

```python
_ = plt.plot([0,1],[0,3])
```

```python
plt.plot([0,1],[0,3], label ='line',color='green')
plt.legend()
plt.xlabel('oś x')
plt.ylabel('oś y')
_ = plt.title('wykres')
```

```python
plt.plot([0,1,2],[0,3,2], label ='line',color='green')
plt.legend()
plt.xlabel('oś x')
plt.ylabel('oś y')
_ = plt.title('wykres')
```

```python
x = np.arange(-3,3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
_ = plt.plot(x, x_cos)
```

### Tworzenie wielu wykresów

```python
plt.subplot(121)
plt.plot([0,1,2],[3,2,5])
plt.subplot(122)
_ = plt.plot([0,1,2],[3,1,4], 'r--')
```

```python
fig=plt.figure(figsize=(12,8)) #rozmiar wykresu w calach

plt.subplot(131)
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([0,1,2],[3,1,4], 'r--')
plt.subplot(133)
_ = plt.plot([0,1,2],[3,1,4], 'ro')
```

### Wykresy słupkowe

```python
_ = plt.bar(x=[0,1,2,3],height=[1.2,3.2,4.5,1.2])
```

```python
_ = plt.bar(x=['a','b','c','d'],height=[1.2,3.2,4.5,1.2]) #dane tekstowe
```

```python
_ = plt.barh(y=['a','b','c','d'],width=[1.2,3.2,4.5,1.2]) #wykres horyzontalny
```

### Wykresy punktowe

```python
x1 = np.random.randn(300)
x2 = np.random.randn(300)

_ = plt.scatter(x1,x2)
```

```python
x3 = np.random.randn(300) * 50 #rozmiar punktów

_ = plt.scatter(x1,x2, s=x3)
```

```python
_ = plt.scatter(x1,x2, s=x3, alpha=0.5) #przezroczystość punktów na wykresie
```
