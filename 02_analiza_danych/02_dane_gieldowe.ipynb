## Obliczanie nowych kolumn

```python
df = fetch_financial_data('UBER')
df.info()
```

```python
df.head()
```

```python
df.describe().T
```

```python
df['Average'] = (df.Open+df.Close)/2
df.head()
```

```python
df = df.sort_index() 
df.head()
```

```python
df[['Close']].shift(1) #przesuniecie danych o 1 wiersz
```

```python
df['Close_shift'] = df.Close.shift(1)
df.head()
```

```python
df['Daily_Change'] = df.Close / df.Close_shift - 1 
df.head()
```

```python
df.Daily_Change.min()
```

```python
df.Daily_Change.max()
```

```python
df.Daily_Change.hist(bins = 50) #histogram zmiennej stopy zwrotu [matplotlib]
```

## Filtrowanie danych

```python
df.head()
```

```python
df_positive = df[df.Daily_Change > 0] #dodatnie wartosci dziennej stopy zwrotu
df_positive
```

```python
df_positive.Daily_Change.mean() #srednia dodatnia stopa zwrotu
```

```python
df_negative = df[df.Daily_Change < 0] #ujemne wartosci dziennej stopy zwrotu
df_negative
```

```python
df_negative.Daily_Change.mean()
```
